### No Show: shouldn't plot

```python
import radis

radis.test_spectrum(verbose=False).plot()
radis.test_spectrum(verbose=False).apply_slit(10, "nm").plot(nfig='same', lw=2)
```

But you can still show it later: 

```python
import matplotlib.pyplot as plt
plt.show()
```

### with show: should plot

```python
import radis

radis.test_spectrum(verbose=False).plot()
radis.test_spectrum(verbose=False).apply_slit(10, "nm").plot(nfig='same', lw=2, show=True)
```

### Equivalent:  use %matplotlib inline

```python
%matplotlib inline
import radis

radis.test_spectrum(verbose=False).plot()
radis.test_spectrum(verbose=False).apply_slit(10, "nm").plot(nfig='same', lw=2)
```
